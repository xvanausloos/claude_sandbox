```python
# Load env variables and create client
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()

client = Anthropic()
model = "claude-haiku-4-5"
```

```python
def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": text
    }
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant",
        "content": text
    }
    messages.append(assistant_message)

def chat(messages):
    # Make a request to the API and print the response
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text
```

```python
# Make a starting list of messages
messages = []

# Add initial user message
add_user_message(messages, "Define quantum computing in one sentence")

# Pass list of messages into chat
answer = chat(messages)

# Take the answer and add it to the list of messages
add_assistant_message(messages, answer)

# Add in the user's follow up question
add_user_message(messages, "Write another sentence")

# Call chat again with the updated list of messages
answer = chat(messages)
answer
```
